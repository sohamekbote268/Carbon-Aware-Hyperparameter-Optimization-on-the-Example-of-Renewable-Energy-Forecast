```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
def preprocess_data(file_path):
    # Read data
    features = pd.read_excel(file_path)
    
    # Filter data
    features = features[(features['MSE'] <= 600) & (features['Sequence Length'] <= 60)]
    features.columns = features.columns.str.replace('WholeDataSet_', '')
    
    # Remove unnecessary columns
    columns_to_remove3 = [col for col in features.columns if 'Part3' in col]
    columns_to_remove2 = [col for col in features.columns if 'Part2' in col]
    columns_to_remove1 = [col for col in features.columns if 'Part1' in col]
    features = features.drop(columns=columns_to_remove1 + columns_to_remove2 + columns_to_remove3)
    
    # Filter rows
    features = features[(features['Tot_IQR'] != 0) & (features['Tot_Std Dev'] != 0)]
    
    # Drop additional columns
    columns_to_remove = ['Tot_IQR', 'Tot_Std Dev', 'Country_Code', 'Value_Count']
    features = features.drop(columns=columns_to_remove)
    
    # One-hot encode the data
    features = pd.get_dummies(features)
    
    return features
```

```python
def train_model(train_features, train_labels):
    # Train Random Forest model
    rf = RandomForestRegressor(n_estimators=1500, random_state=42)
    rf.fit(train_features, train_labels)
    return rf
```

```python
def predict_data(model, test_features):
    # Predict using the trained model
    predictions = model.predict(test_features)
    return predictions
```

```python
def merge_data(predictions, features1_first_elements, labels, features1):
    # Merge prediction results with additional data
    result_df = pd.DataFrame({'WholeDataSet_STD_DEV': features1_first_elements, 'Features': features1, 'Actual Values': labels, 'Predicted Values': predictions})
    return result_df
```

```python
def plot_results(result_df, time_series_features):
    # Merge additional data
    merged_df = pd.merge(result_df, time_series_features, on='WholeDataSet_STD_DEV', how='inner')
    
    # Save merged dataframe
    merged_df.to_excel('predictions3.xlsx', index=False)
    
    # Read the saved data
    df = pd.read_excel('predictions3.xlsx')
    
    # Group data by country code
    grouped = df.groupby('Country_Code')
    
    # Initialize lists
    common_features_list = []
    country_codes_list = []
    
    # Find common features
    for country_code, group in grouped:
        sorted_actual = group.sort_values('Actual Values').head(4)
        sorted_predicted = group.sort_values('Predicted Values').head(4)
        actual_features = set(sorted_actual['Features'])
        predicted_features = set(sorted_predicted['Features'])
        common_features = actual_features.intersection(predicted_features)
        common_features_list.append(len(common_features))
        country_codes_list.append(country_code)
    
    # Determine colors
    colors = []
    for country_code, group in grouped:
        sorted_actual = group.sort_values('Actual Values').head(1)
        sorted_predicted = group.sort_values('Predicted Values').head(1)
        actual_feature = sorted_actual['Features'].iloc[0]
        predicted_feature = sorted_predicted['Features'].iloc[0]
        colors.append('skyblue' if actual_feature in predicted_feature else 'skyblue')
    
    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.bar(country_codes_list, common_features_list, color=colors)
    plt.xlabel('Country Code')
    plt.ylabel('Number of Common Features')
    plt.title('Common Features between Lowest 4 Actual and Predicted MSE Values by Country')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    for country_code, group in grouped:
        print("Country Code:", country_code)

        # Sort the group by "Actual Values" and "Predicted Values" separately and select the lowest 4 values
        sorted_actual = group.sort_values('Actual Values').head(4)
        sorted_predicted = group.sort_values('Predicted Values').head(4)

        # Extract the features for the lowest 4 actual and predicted values
        actual_features = set(sorted_actual['Features'])
        predicted_features = set(sorted_predicted['Features'])

        # Find the intersection of features between the lowest 4 actual and predicted values
        common_features = actual_features.intersection(predicted_features)

        # Print the common features along with the lowest values
        print("Common Features:")
        print(common_features)
        print("Top 4 Lowest Actual Values:")
        print(sorted_actual[['Features', 'Actual Values']])
        print("Top 4 Lowest Predicted Values:")
        print(sorted_predicted[['Features', 'Predicted Values']])
        print("=" * 50)
```

```python
def main():
    # Preprocess and train on carbon intensity data
    features = preprocess_data(r'C:\Users\soham\Soham Master Thesis\Datasets\Generated_datasets_for_Regressor_Model\carbon_intensity_data_merged.xlsx')
    labels = np.array(features['MSE'])
    features = features.drop('MSE', axis=1)
    feature_list = list(features.columns)
    features = np.array(features)
    train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=0.20, random_state=42)
    rf_model = train_model(train_features, train_labels)
    
    # Preprocess and predict on percentage renewable data
    features1 = preprocess_data(r'C:\Users\soham\Soham Master Thesis\Datasets\Generated_datasets_for_Regressor_Model\percentage_renwable_data_merged.xlsx')
    labels = np.array(features1['MSE'])
    features1 = features1.drop('MSE', axis=1)
    feature_list = list(features1.columns)
    features1 = np.array(features1)
    predictions = predict_data(rf_model, features1)
    
    # Merge and plot results
    features1 = features1.tolist()
    features1_first_elements = [sublist[0] for sublist in features1]
    features1 = [inner_list[-3:-2] + inner_list[-2:] for inner_list in features1]
    for inner_list in features1:
        if inner_list[1] == 1:
            inner_list[1] = "MinMaxScaler"
        elif inner_list[1] == 0:
            inner_list[1] = "StandardScaler"
    features1 = [[inner_list[0], inner_list[1]] for inner_list in features1]
    
    result_df = merge_data(predictions, features1_first_elements, labels, features1)
    
    time_series_features = pd.read_excel(r'C:\Users\soham\Soham Master Thesis\Datasets\Generated_time_series_data\all_features_output_percent_renewable.xlsx')
    time_series_features = time_series_features[['WholeDataSet_STD_DEV', 'Country_Code']]
    
    plot_results(result_df, time_series_features)

if __name__ == "__main__":
    main()
```
